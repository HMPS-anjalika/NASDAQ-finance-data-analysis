```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Step 2: Load the Datasets
meta_stock_data = pd.read_csv('data/stocks/META.csv') 
usd_inr_data = pd.read_csv('data/currencies/INR.csv')
bitcoin_data = pd.read_csv('data/cryptocurrencies/BTC-USD.csv')
```

```python
## Inspecting Columns
print("Meta Stock Data Columns:", meta_stock_data.columns)
print("USD/INR Data Columns:", usd_inr_data.columns)
print("Bitcoin Data Columns:", bitcoin_data.columns)
```

```python
# Display the first few rows of each dataset
print("Meta Stock Data:")
print(meta_stock_data.head(), "\n")

print("USD/INR Exchange Rate Data:")
print(usd_inr_data.head(), "\n")

print("Bitcoin Data:")
print(bitcoin_data.head(), "\n")
```

```python
# Convert 'Date' columns to datetime format where available
meta_stock_data['Date'] = pd.to_datetime(meta_stock_data['Date'])

# Add a 'Date' column to USD/INR and Bitcoin data using 'Year'
usd_inr_data['Date'] = pd.to_datetime(usd_inr_data['Year'].astype(str) + '-01-01')
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Year'].astype(str) + '-01-01')

# Handle missing values using the updated method
meta_stock_data = meta_stock_data.ffill()
usd_inr_data = usd_inr_data.ffill()
bitcoin_data = bitcoin_data.ffill()
```

```python
# Display the first few rows of each dataset
print("Meta Stock Data:")
print(meta_stock_data.head(), "\n")

print("USD/INR Exchange Rate Data:")
print(usd_inr_data.head(), "\n")

print("Bitcoin Data:")
print(bitcoin_data.head(), "\n")
```

```python
# Resample Apple stock data to yearly frequency
meta_stock_data.set_index('Date', inplace=True)
meta_stock_yearly = meta_stock_data.resample('YE').last().reset_index()
```

```python
print("Meta Stock Yearly Data:\n", meta_stock_yearly.head())
```

```python
print("Meta Stock Yearly Date Range:", meta_stock_yearly['Date'].min(), "to", meta_stock_yearly['Date'].max())
print("USD/INR Date Range:", usd_inr_data['Date'].min(), "to", usd_inr_data['Date'].max())
print("Bitcoin Date Range:", bitcoin_data['Date'].min(), "to", bitcoin_data['Date'].max())
```

```python
# Manually set year-start dates
meta_stock_yearly['Date'] = pd.to_datetime(meta_stock_yearly['Date'].dt.year.astype(str) + '-01-01')
```

```python
print("Meta Stock Yearly Data (Updated Dates):\n", meta_stock_yearly.head())
```

```python
# Define the overlapping period
start_date = '2014-01-01'
end_date = '2023-01-01'

# Filter datasets to overlapping period
meta_stock_filtered = meta_stock_yearly[(meta_stock_yearly['Date'] >= start_date) & (meta_stock_yearly['Date'] <= end_date)]
usd_inr_filtered = usd_inr_data[(usd_inr_data['Date'] >= start_date) & (usd_inr_data['Date'] <= end_date)]
bitcoin_filtered = bitcoin_data[(bitcoin_data['Date'] >= start_date) & (bitcoin_data['Date'] <= end_date)]
```

```python
print("Meta Stock Filtered Date Range:", meta_stock_filtered['Date'].min(), "to", meta_stock_filtered['Date'].max())
print("USD/INR Filtered Date Range:", usd_inr_filtered['Date'].min(), "to", usd_inr_filtered['Date'].max())
print("Bitcoin Filtered Date Range:", bitcoin_filtered['Date'].min(), "to", bitcoin_filtered['Date'].max())
```

```python
# Merge datasets
merged_data = meta_stock_filtered[['Date', 'Close']].merge(
    usd_inr_filtered[['Date', 'Close']], on='Date', suffixes=('_META', '_USDINR'), how='inner'
).merge(
    bitcoin_filtered[['Date', 'Close']], on='Date', how='inner'
)
merged_data.rename(columns={'Close': 'BTC_Close'}, inplace=True)

# Check merged data
print("Merged Data Shape:", merged_data.shape)
print("Merged Data Sample:\n", merged_data.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
merged_data[['Close_META', 'Close_USDINR', 'BTC_Close']] = scaler.fit_transform(
    merged_data[['Close_META', 'Close_USDINR', 'BTC_Close']]
)

print("Normalized Data:\n", merged_data.head())
```

```python
print("Summary Statistics:\n", merged_data.describe())
```

```python
correlation_matrix = merged_data[['Close_META', 'Close_USDINR', 'BTC_Close']].corr()
print("Correlation Matrix:\n", correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot historical trends
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Date'], merged_data['Close_META'], label='Meta Stock Price')
plt.plot(merged_data['Date'], merged_data['Close_USDINR'], label='USD/INR Exchange Rate')
plt.plot(merged_data['Date'], merged_data['BTC_Close'], label='Bitcoin Price')
plt.xlabel('Date')
plt.ylabel('Normalized Value')
plt.title('Historical Trends: Meta Stock, USD/INR, and Bitcoin Prices')
plt.legend()
plt.show()

# Plot correlation heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Prepare data for regression
X = merged_data[['Close_USDINR', 'BTC_Close']]
y = merged_data['Close_META']

# Fit regression model
regressor = LinearRegression()
regressor.fit(X, y)

# Print regression coefficients
print("Regression Coefficients:", regressor.coef_)
```
